```python
import pandas as pd
import numpy as np
%pylab inline
```

```python
data = pd.read_csv("./sample_data/california_housing_train.csv")
data.head()
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = data.drop("median_house_value", axis=1)
y = data['median_house_value']

Xtr, Xval, ytr, yval = train_test_split(X, y, test_size=0.5, random_state=0)

#arvere = DecisionTreeRegressor(max_depth=8, random_state=0)
arvere = DecisionTreeRegressor(min_samples_leaf=21, random_state=0)
arvere.fit(Xtr, ytr)

p = arvere.predict(Xval)

np.sqrt(mean_squared_error(yval, p))
```

```python
# max_depth=x, é a profundidade dos nós
# min_samples_leaf = x, nesse caso, é a quantidade de amostras por nós, mario filho usa esse daqui.
```
